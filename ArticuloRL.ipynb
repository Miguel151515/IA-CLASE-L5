### Se puede visualizar por ejemplo cantidad de palabras vs compartidos

```python
# Visualizar datos de entrada
colores = ['orange','blue']
tama√±os = [30,60]

f1 = data['Word count'].values
f2 = data['# Shares'].values

# Pintamos en dos colores los puntos por debajo de la media de cantidad de palabras
asignar = []
for index, row in data.iterrows():
    if(row['Word count']>1808):
        asignar.append(colores[0])
    else:
        asignar.append(colores[1])
        
plt.scatter(f1, f2, c=asignar, s=tama√±os[0])
plt.show()
```

```python
# VISUALIZAR los datos en la zona donde se concentran m√°s los puntos
# esto es en el eje X: entre 0 y 3.500
# y en el eje Y: entre 0 y 80.000
filtered_data = data[(data['Word count'] <= 3500) & (data['# Shares'] <= 80000)]

f1 = filtered_data['Word count'].values
f2 = filtered_data['# Shares'].values

# Pintar en colores los puntos por debajo y por encima de la media de Cantidad de Palabras
asignar=[]
for index, row in filtered_data.iterrows():
    if(row['Word count']>1808):
        asignar.append(colores[0])
    else:
        asignar.append(colores[1])
    
plt.scatter(f1, f2, c=asignar, s=tama√±os[0])
plt.show()
```

```python
# Se revisan los cambios de valores una vez filtrados
filtered_data.describe()
```

### Regresion Lineal Simple (1 atributo con Phyton y SKLearn

```python
 # Asignamos el atributo de entrada X para entrenamiento y la etiqueta Y
dataX = filtered_data[['Word count']]
x_train = np.array(dataX)
y_train = filtered_data['# Shares'].values
```

```python
# Se genera el objeto de regresion lineal
regr = linear_model.LinearRegression()

# Se encuentra el modelo
regr.fit(x_train, y_train)

# Se realizan las predicciones
y_pred = regr.predict(x_train)

# Se revisan los coeficientes obtenidos. En este caso la tangente
print('Coeficientes: \n', regr.coef_)

# Este es el valor de donde corta el eje Y (x=0)
print('Termino Independiente: \n', regr.intercept_)

# Error cuadrado medio
print('Media del error cuadrado %.2f' % mean_squared_error(y_train, y_pred))

# Puntuaje de varianza. El mejor puntuaje es 1.0
print('Puntuaje de la varianza: %2.f' % r2_score(y_train, y_pred))
```

De la ecuacion de la recta y = mx + b la pendiente 'm' es el coeficiente 5.69765366 y el termino independiente 'b' es 11200.303223074163. Existe un error cuadratico grande; por lo cual este modelo no es el mas adecuado. Pero estamos aprendiendo su funcionamiento, lo que resulta importante por el momento. Esto lo vemos reflejado en el puntuaje de varianza el cual debe ser cerca de 1.0

```python
plt.scatter(x_train[:,0], y_train, c=asignar, s=tama√±os[0])
plt.plot(x_train[:,0], y_pred, color='red', linewidth = 3)

plt.xlabel('Cantidad de Palabras')
plt.ylabel('Cantidad de veces que se compartio')
plt.title('Regresion lineal')

plt.show()
```

### Prediccion 1

```python
# Predecir
# Cuantos "Shares" se obtendra de un articulo de 2000 palabras
y_dosMil = regr.predict([[2000]])
print('Cantidad predecida: ', int(y_dosMil))
```
